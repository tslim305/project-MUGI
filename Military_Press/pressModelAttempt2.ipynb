```python
cap = cv2.VideoCapture('HigherResCurlVsShoulderPress/batch_ID_df3a86c8-fd1d-4d4e-beb6-0a261038715c/3720fdab-1aa9-4ba9-9de1-b8ae7193da7f/video.mp4')

frame_list = []

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            print("the frame is empty")
            
        #recoloring feed to make it passable to mediapipe
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False



        results = holistic.process(image)


        try:
            pose = results.pose_landmarks.landmark
            pose_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in pose]).flatten())

            #redudent detection of face landmarks....may come in use later specifically for detecting strikes to the dome
            face = results.face_landmarks.landmark
            face_row = list(np.array([[landmark.x, landmark.y, landmark.z, landmark.visibility] for landmark in face]).flatten())

            row = pose_row+face_row

            row.insert(0, class_name)

            with open('coords.csv', mode='a', newline='') as f:
                csv_writer = csv.writer(f, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csv_writer.writerow(row)

        except :
            pass

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break


cap.release()
cv2.destroyAllWindows()

```

<big><strong>CREATING CUSTOMER CLASSIFICATION MODEL TO CLASSIFY FROM KEYPOINTS GIVEN </big></strong>

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('coords.csv')

df.head(25)
```

```python
#filtering to only get rows where the class column value is bicep curl
df[df['class']=='bicep_curl']
```

```python
X = df.drop('class', axis=1) #here we drop the class column so we are extracting our feautures i.e the coordinates to the keypoints
y = df['class'] # the target values we want to identify / match with X
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=1234)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
'''
Think of a pipeline as a streamlined way to setup all the preproccessing of data
that has be done before being fed to any specifed model

the actual pipeline varible is a dictionary where the abbriviations are keys that point
to the models themselves.

So in this case each model first normalizees the data(standardscalar) then passes it to its respective model 

'''

pipelines = {

    'lr':make_pipeline(StandardScaler(), LogisticRegression()),
    'rc':make_pipeline(StandardScaler(), RidgeClassifier()),
    'rf':make_pipeline(StandardScaler(), RandomForestClassifier()),
    'gb':make_pipeline(StandardScaler(), GradientBoostingClassifier()),
    
}

list(pipelines.values())[1]
```

```python
''' Here we cycle through all the possible models we have and go through their indavidual pipelines'''

fit_models = {}
for algo, pipeline in pipelines.items():
    model = pipeline.fit(X_train, y_train)
    fit_models[algo] = model

#this is where you stopped; this has not ran
```

```python
fit_models
```

<strong><big>Evaluating each pipeline to choose the best</strong></big>

```python
fit_models['rf'].predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import pickle

'''pickle is used to save a model down to a disc. keep in mind for app. this will help when moving model to app '''
```

```python
for algo, model in fit_models.items():
    yhat = model.predict(X_test)
    print(algo, accuracy_score(y_test, yhat))


''' yhat is to store prediction; common in data science'''
```
